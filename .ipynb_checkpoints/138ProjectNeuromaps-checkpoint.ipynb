```python
import neuromaps
from neuromaps import datasets
import nibabel as nib
from neuromaps.datasets import fetch_annotation
from neuromaps import transforms
from neuromaps import images, nulls, resampling
import abagen
from neuromaps.parcellate import Parcellater
from neuromaps import resampling
from neuromaps import stats
from neuromaps import plotting
import numpy as np
```

### Alzheimer's Disease Association Map - Neurosynth

```python
# load ad association map
AD_association = nib.load('neurosynth_maps/alzheimer_association-test_z_FDR_0.01.nii')
```

```python
# transform AD map from mni to fsavg
fsavg_AD = transforms.mni152_to_fsaverage(AD_association, '10k')
```

```python
AD_surf_plot = plotting.plot_surf_template(fsavg_AD, "fsaverage", "10k")
```

### Gene Expression - Neuromaps/Abagen

```python
genepc = datasets.fetch_annotation(desc='genepc1', return_single=True)
```

```python
genepc_plot = plotting.plot_surf_template(genepc, "fsaverage", '10k')
```

```python
corr1 = stats.compare_images(fsavg_AD, genepc)
print(f'Correlation: r = {corr1:.02f}')
```

```python
rotated = nulls.alexander_bloch(fsavg_AD, atlas='fsaverage', density='10k',n_perm=100, seed=1234)
print(rotated.shape)
```

```python
corr, pval = stats.compare_images(fsavg_AD, genepc, nulls=rotated)
print(f'r = {corr:.3f}, p = {pval:.3f}')
```
